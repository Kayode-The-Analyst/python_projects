```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\User\Documents\Kayode_Omotayo\Data Analysis\Python_Projects\Customer Call List.xlsx')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip('..._/')
df
```

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('\W', '', regex=True)
df
```

```python

df=df.replace('N/a', "")
```

```python
df
```

```python
df=df.replace('Na', "")
```

```python
df
```

```python
df=df.fillna("")
df
```

```python
df["Paying Customer"]= df["Paying Customer"].replace('Yes', 'Y')
df["Paying Customer"]= df["Paying Customer"].replace('No', 'N')
```

```python

df["Do_Not_Contact"]= df["Do_Not_Contact"].replace('Yes', 'Y')
df["Do_Not_Contact"]= df["Do_Not_Contact"].replace('No', 'N')
df

```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x : x[0:3]+ "-" + x[3:6] + "-" + x[6:9])
df
```

```python
df[["Street","City","Post_Code"]]=df["Address"].str.split(',', expand=True)
df
```

```python
for x in df.index:
     if df.loc[x, "Do_Not_Contact"]=='Y':
         df.drop(x, inplace=True)
df
```

```python
for x in df.index:
     if df.loc[x, "Do_Not_Contact"]=='':
         df.drop(x, inplace=True)
df
```

```python
df=df.reset_index(drop=True)
df
```

```python
for x in df.index:
     if df.loc[x, "Phone_Number"]=='--':
         df.drop(x, inplace=True)
df
```
